```python
import pandas as pd
import json 
import requests as req 
import apikeys
from citipy import citipy
import matplotlib.pyplot as plt
import numpy as np
import random
import datetime as dt
```

```python
url = 'http://api.openweathermap.org/data/2.5/weather' 
api_key = apikeys.OWM_API_KEY
```

```python
cities = []
cit = []
clouds = []
country = []
date = []
humidity = []
lat = []
lon = []
maxTemp = []
windSpeed = []
```

```python
cityCount = 0
while cityCount < 550:
    city1 = citipy.nearest_city(random.randint(-90,90), random.randint(-180,180))
    city2 = city1.city_name
    
    params = {'appid': apikeys.OWM_API_KEY, 'q': city2, 'units':'imperial'}
    weather_response = req.get(url, params=params)
    
    weather_json = weather_response.json()

    if weather_json['cod'] == 200:
        cities.append(city2)
        
        query_url = url + '?appid=' + api_key + '&q=' + city2
        print("Processing Record: " + str(cityCount +1) + "|" + city2 + ", " + weather_json['sys']['country']
             + '\n' + query_url)
    
        cit.append(weather_json['name'])
        clouds.append(weather_json['clouds']['all'])
        country.append(weather_json['sys']['country'])
        date.append(weather_json['dt'])
        humidity.append(weather_json['main']['humidity'])
        lat.append(weather_json['coord']['lat'])
        lon.append(weather_json['coord']['lon'])
        maxTemp.append(weather_json['main']['temp_max'])
        windSpeed.append(weather_json['wind']['speed'])
    
        cityCount = cityCount + 1
```

```python
weatherDF = pd.DataFrame({'city':cit, 'clouds':clouds, 'country':country, 'date':date, 'humidity':humidity,
                          'latitude':lat, 'longtitude':lon, 'Max Temp':maxTemp, 'Wind Speed':windSpeed})
weatherDF = weatherDF[['city', 'clouds', 'country', 'date', 'humidity', 'latitude', 'longtitude', 'Max Temp', 'Wind Speed']]
weatherDF.info()
```

```python
weatherDF.to_csv('weatherData.csv')
```

```python
plt.scatter(weatherDF['latitude'], weatherDF['Max Temp'], marker="o", 
            facecolors="blue", edgecolors="black", alpha=0.75)
```

```python
plt.title("City Latitude vs. Max. Temperature" + " (" + dt.datetime.today().strftime("%m/%d/%Y") + ")")
plt.xlabel("Latitude")
plt.ylabel("Max. Temperature (F)")
plt.xlim(-100, 100)
plt.ylim(-60, 120)
plt.grid()
```
